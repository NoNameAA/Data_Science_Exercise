```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
samples = 500
truth_value_x = np.linspace(0, 6, samples)
truth_value_y = np.sin(truth_value_x)

random_num = np.random.uniform(-0.3, 0.3, samples)
noise_value_y = truth_value_y + random_num

plt.plot(truth_value_x, truth_value_y, 'r-', linewidth = 4)
plt.plot(truth_value_x, noise_value_y, 'b.', alpha = 0.5)
plt.legend(['Truth Value', 'Sensor reading'])
plt.xlabel('Time')
plt.ylabel('Value of thing we care about')
plt.show()
```

```python
import statsmodels.api as sm
lowess = sm.nonparametric.lowess
filtered = lowess(noise_value_y, truth_value_x, frac=0.1)
plt.plot(truth_value_x, noise_value_y, 'b.', alpha=0.5)
plt.plot(filtered[:, 0], filtered[:, 1], 'r-', linewidth=4)
plt.legend(['Sensor readings', 'Reconstrcted signal'])
plt.show()
```
